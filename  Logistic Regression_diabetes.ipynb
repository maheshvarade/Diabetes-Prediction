```python
### here we seen that same of columns has outlayer so 

from scipy import stats

# Function to remove outliers using Z-score
def remove_outliers_zscore(df, threshold=5):
    z_scores = stats.zscore(df.select_dtypes(include=['number']))
    mask = (abs(z_scores) < threshold).all(axis=1)  # Keep rows where all columns are within threshold
    return df[mask]

filtered_data = remove_outliers_zscore(data)
print(filtered_data)
```

```python
filtered_data.describe()
```

# split the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=45)
```

 # LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

## imbalanced dataset (e.g., 95% class 0, 5% class 1) can cause the model to ignore the minority class.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(x_train, y_train)
```

```python
RF.fit(X_resampled, y_resampled)
pred_smote=RF.predict(x_test)
```

```python
accuracy_score(pred_smote,y_test)*100
```
